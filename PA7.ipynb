```python
import math

class calculator():
    def __init__(self,input1,input2):
        self.input1=input1
        self.input2=input2
    def addition(self):
        return self.input1+self.input2
    def subtraction(self):
        return self.input1-self.input2
    def multiplication(self):
        return self.input1*self.input2
    def division(self):
        return self.input1/self.input2
    def Square(self): #5
        return self.input1**2
    def Cube(self):
        return self.input1**3
    def Exponent(self):
        return self.input1**self.input2
    def Square_Root(self):
        return math.sqrt(self.input1)
    def Cube_Root(self):
        return math.cbrt(self.input1)
    def Factorial(self):#10
        return math.factorial(self.input1)
    def sin(self):
        return math.sin(self.input1)
    def Cos(self):
        return math.cos(self.input1)
    def Tan(self):
        return math.tan(self.input1)
    def log(self):
        return math.log(self.input1,self.input2)
    def log10(self):#15
        return math.log12(self.input1)
    

choice=1
z=3
while choice!=0:
    print()
    print("0. Exit")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Square")
    print("6. Cube")
    print("7. Exponent")
    print("8. Square Root")
    print("9. Cube Root")
    print("10. Factorial")
    print("11. Sin")
    print("12. Cos")
    print("13. Tan")
    print("14. log")
    print("15. log10")
    choice=int(input("Enter your choice...."))
    
    if z==3:
        if choice in [1,2,3,4,7,]:
            input_1=int(input("Enter first number: "))
            input_2=int(input("Enter second number: "))
            my_calc=calculator(input_1,input_2)
            
        elif choice in [5,6,8,9,10,11,12,13,14,15]:
            input_1=int(input("Enter a number: "))
            my_calc=calculator(input_1,0)
            
    
    if choice==1:
        print(f"Result: {my_calc.addition()}")
        
    elif choice==2:
        print(f"Result: {my_calc.subtraction()}")
       
    elif choice==3:
        print(f"Result: {my_calc.multiplication()}")
        
    elif choice==4:
        print(f"Result: {my_calc.division()}")
        
    elif choice==5:
        print(f"Result: {my_calc.Square()}")
        
    elif choice==6:
        print(f"Result: {my_calc.Cube()}")
        
    elif choice==7:
        print(f"Result: {my_calc.Exponent()}")
        
    elif choice==8:
        print(f"Result: {my_calc.Square_Root()}")
        
    elif choice==9:
        print(f"Result: {my_calc.Cube_Root()}")
        
    elif choice==10:
        print(f"Result: {my_calc.Factorial()}")
        
    elif choice==11:
        print(f"Result: {my_calc.sin()}")
        
    elif choice==12:
        print(f"Result: {my_calc.Cos()}")
        
    elif choice==13:
        print(f"Result: {my_calc.Tan()}")
        
    elif choice==14:
        print(f"Result: {my_calc.log()}")
        
    elif choice==15:
        print(f"Result: {my_calc.log10()}")
        
    elif choice==0:
        print("Exit")
    else:
        print("Sorry this is not an option... pick again")
```
